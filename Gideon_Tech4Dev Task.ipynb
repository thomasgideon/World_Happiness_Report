### Building Linear Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import RobustScaler
```

```python
robust_scaler = RobustScaler()
X = robust_scaler.fit_transform(X) #scaling the features
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
print("Training_score : " , linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

print("Testing_score : ", r2_score(y_test, y_pred))
```

Our model returned R-Squared = 76.2%. This means that about 76% variation in nations' happiness is explained by the predictors.

```python
# MULTIPLE LR

#We can also use statsmodels to predict the Happiness Score.

import statsmodels.api as sm
X = data_combined[['Economics', 'Social Support', 'Life Expectancy', 'Freedom', 'Absence of Corruption', 'Generosity']]
Y = data_combined['Happiness Score']

X_sm = X = sm.add_constant(X)
model = sm.OLS(Y,X_sm)
model.fit().summary()
```

The OLS Summary above reveals that all the features are significant positive predictors or contributors to nations' happiness, even at 99% confidence level. However, the most contributing factors or features are `Freedom`(1.4796), `Economics`(1.1477), `Life Expectancy`(1.0033), `Absence of Corruption`(0.8510), `Social Support`(0.6406), and `Generosity`(0.5941), in that other.  

For example, the above result means that a unit increase or change in people's freedom will cause or result in an increase in people's happiness by 1.4796 units. Other interpreations follow in the same way.
